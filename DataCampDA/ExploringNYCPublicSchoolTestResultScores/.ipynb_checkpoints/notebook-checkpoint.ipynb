## Step 2: Feature Understanding

>  Which NYC schools have the best math results?

```python
best_math_schools = schools.loc[schools['average_math'] >= (800 * 0.8)][['school_name','average_math']]
best_math_schools = best_math_schools.sort_values(by='average_math', ascending=False)
```

> What are the top 10 performing schools based on the combined SAT scores?

```python
schools.head()
```

```python
schools['total_SAT'] = schools[['average_math','average_reading','average_writing']].apply(lambda x: x[0] + x[1] + x[2], axis=1)
schools
```

OR

```python
schools['total_SAT'] = schools[['average_math','average_reading','average_writing']].sum(axis=1)
schools
```

```python
top_10_schools = schools.sort_values(by='total_SAT', ascending=False)[['school_name','total_SAT']].reset_index(drop=True).head(10)
top_10_schools
```

> Which single borough has the largest standard deviation in the combined SAT score?

```python
largest_std_dev = schools.pivot_table(values='total_SAT', index='borough', aggfunc=["std","mean","count"]).reset_index()
largest_std_dev.columns = ['BOROUGH','STD','MEAN','COUNT']
largest_std_dev = largest_std_dev.loc[largest_std_dev['STD'] == largest_std_dev['STD'].max()]
largest_std_dev.columns = ['borough', 'std_SAT','average_SAT','num_schools']
largest_std_dev_numeric_cols = largest_std_dev.select_dtypes(include=['int','float']).columns
largest_std_dev[largest_std_dev_numeric_cols] = largest_std_dev[largest_std_dev_numeric_cols].round(2)

largest_std_dev
```
